```python
 #solve missing-values problem:
import numpy as np
import pandas as pd
data=pd.read_csv('DataSet2.csv')
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(data.iloc[:,1:3])
data.iloc[:,1:3]= imputer.transform(data.iloc[:,1:3])
data
```

```python
#solve alphapet problem:
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([('encoder', OneHotEncoder(), [0])], remainder='passthrough')
data = pd.DataFrame(ct.fit_transform(data))
data
```

```python
#label-transformation
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data.iloc[:,-1] = le.fit_transform(data.iloc[:,-1])
data
```

```python
#standarlization-for-data
from sklearn.preprocessing import StandardScaler 
st= StandardScaler(copy=True,with_mean=True,with_std=True)  
data= pd.DataFrame(st.fit_transform(data))  
data
```

```python
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data= pd.DataFrame(scaler.fit_transform(data)) 
data
```
